```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures

# Load and preprocess data
file_path = r"D:\B TECH COMPUTER SCIENCE AND ENGINEERING\SEMESTER 05\MACHINE LEARNING\ML PROJECT\EMBEDDINGS_GPT2.xlsx"
data = pd.read_excel(file_path)
X = data.iloc[:, 0:768]
y = data['Average_Marks']

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define models
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(),
    "Lasso Regression": Lasso(),
    "ElasticNet Regression": ElasticNet(),
    "SVR": SVR(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(),
    "Gradient Boosting": GradientBoostingRegressor(),
    "K-Nearest Neighbors": KNeighborsRegressor(),
    "Polynomial Regression (2nd Degree)": make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
}

# Initialize lists to store results
model_names = []
avg_rmse_scores = []
avg_r2_scores = []

# Perform 10-fold cross-validation for each model
for name, model in models.items():
    # Cross-validation for RMSE
    neg_mse_scorer = make_scorer(mean_squared_error, greater_is_better=False)
    cv_scores_mse = cross_val_score(model, X_scaled, y, cv=10, scoring=neg_mse_scorer)
    cv_rmse_scores = np.sqrt(-cv_scores_mse)  # Convert negative MSE to positive RMSE

    # Cross-validation for RÂ²
    cv_scores_r2 = cross_val_score(model, X_scaled, y, cv=10, scoring='r2')

    # Calculate and store average RMSE and RÂ²
    avg_rmse = np.mean(cv_rmse_scores)
    avg_r2 = np.mean(cv_scores_r2)
    model_names.append(name)
    avg_rmse_scores.append(avg_rmse)
    avg_r2_scores.append(avg_r2)

# Create a DataFrame to hold results
cv_results_df = pd.DataFrame({
    "Model": model_names,
    "Avg RMSE (10-fold)": avg_rmse_scores,
    "Avg RÂ² (10-fold)": avg_r2_scores
})

# Display the cross-validation performance DataFrame
print(cv_results_df)

# Plotting RMSE and RÂ² scores for each model
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# RMSE Bar Plot
axes[0].barh(model_names, avg_rmse_scores, color='skyblue', edgecolor='black')
axes[0].set_xlabel('Average RMSE (10-fold)')
axes[0].set_title('10-Fold Cross-Validation RMSE for Different Regression Algorithms')

# RÂ² Bar Plot
axes[1].barh(model_names, avg_r2_scores, color='lightgreen', edgecolor='black')
axes[1].set_xlabel('Average RÂ² (10-fold)')
axes[1].set_title('10-Fold Cross-Validation RÂ² for Different Regression Algorithms')

plt.tight_layout()
plt.show()
```
