```python
# 01 Linear Regression
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Load data
file_path = r"D:\B TECH COMPUTER SCIENCE AND ENGINEERING\SEMESTER 05\MACHINE LEARNING\ML PROJECT\EMBEDDINGS_GPT2.xlsx"
data = pd.read_excel(file_path)

# Features and target variable
X = data.iloc[:, 0:768]
y = data['Average_Marks']

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Model and threshold
model = LinearRegression()
threshold = 50

# Train and predict
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Regression metrics
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, y_pred)

# Convert predictions to binary classes based on threshold
y_pred_classes = (y_pred >= threshold).astype(int)
y_test_classes = (y_test >= threshold).astype(int)

# Classification metrics
accuracy = accuracy_score(y_test_classes, y_pred_classes)
precision = precision_score(y_test_classes, y_pred_classes, zero_division=0)
recall = recall_score(y_test_classes, y_pred_classes, zero_division=0)
f1 = f1_score(y_test_classes, y_pred_classes, zero_division=0)
conf_matrix = confusion_matrix(y_test_classes, y_pred_classes)

# Display regression metrics
print("Linear Regression Performance:")
print(f"MSE: {mse}, RMSE: {rmse}, RÂ²: {r2}")
print(f"Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1 Score: {f1}")

# Plot confusion matrix with a colorful palette
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="YlGnBu", cbar=True, annot_kws={"size": 14},
            xticklabels=["Predicted Negative", "Predicted Positive"], 
            yticklabels=["Actual Negative", "Actual Positive"],
            linewidths=1, linecolor='black', square=True)

plt.xlabel("Predicted Class")
plt.ylabel("Actual Class")
plt.title("Confusion Matrix with Colors")
plt.show()

# Plot classification metrics
classification_metrics = {
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall,
    "F1 Score": f1
}

# Convert metrics dictionary to DataFrame for plotting
metrics_df = pd.DataFrame(list(classification_metrics.items()), columns=["Metric", "Score"])

# Plot the classification metrics
plt.figure(figsize=(8, 5))
sns.barplot(data=metrics_df, x="Metric", y="Score", palette="coolwarm")
plt.title("Classification Metrics for Linear Regression")
plt.ylabel("Score")
plt.ylim(0, 1)  # Assuming scores are within 0-1
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# 08 K-Nearest Neighbors Regression
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Load data
file_path = r"D:\B TECH COMPUTER SCIENCE AND ENGINEERING\SEMESTER 05\MACHINE LEARNING\ML PROJECT\EMBEDDINGS_GPT2.xlsx"
data = pd.read_excel(file_path)

# Features and target variable
X = data.iloc[:, 0:768]
y = data['Average_Marks']

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Model and threshold
model = KNeighborsRegressor()
threshold = 50

# Train and predict
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Regression metrics
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, y_pred)

# Convert predictions to binary classes based on threshold
y_pred_classes = (y_pred >= threshold).astype(int)
y_test_classes = (y_test >= threshold).astype(int)

# Classification metrics
accuracy = accuracy_score(y_test_classes, y_pred_classes)
precision = precision_score(y_test_classes, y_pred_classes, zero_division=0)
recall = recall_score(y_test_classes, y_pred_classes, zero_division=0)
f1 = f1_score(y_test_classes, y_pred_classes, zero_division=0)
conf_matrix = confusion_matrix(y_test_classes, y_pred_classes)

# Display regression metrics
print(" K-Nearest Neighbors Regression Performance:")
print(f"MSE: {mse}, RMSE: {rmse}, RÂ²: {r2}")
print(f"Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1 Score: {f1}")

# Plot confusion matrix with a colorful palette
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="YlGnBu", cbar=True, annot_kws={"size": 14},
            xticklabels=["Predicted Negative", "Predicted Positive"], 
            yticklabels=["Actual Negative", "Actual Positive"],
            linewidths=1, linecolor='black', square=True)

plt.xlabel("Predicted Class")
plt.ylabel("Actual Class")
plt.title("Confusion Matrix with Colors")
plt.show()

# Plot classification metrics
classification_metrics = {
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall,
    "F1 Score": f1                                             
}

# Convert metrics dictionary to DataFrame for plotting
metrics_df = pd.DataFrame(list(classification_metrics.items()), columns=["Metric", "Score"])

# Plot the classification metrics
plt.figure(figsize=(8, 5))
sns.barplot(data=metrics_df, x="Metric", y="Score", palette="coolwarm")
plt.title("Classification Metrics for K-Nearest Neighbors Regression")
plt.ylabel("Score")
plt.ylim(0, 1)  # Assuming scores are within 0-1
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 09 Polynomial Regression (2nd Degree)
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Load data
file_path = r"D:\B TECH COMPUTER SCIENCE AND ENGINEERING\SEMESTER 05\MACHINE LEARNING\ML PROJECT\EMBEDDINGS_GPT2.xlsx"
data = pd.read_excel(file_path)

# Features and target variable
X = data.iloc[:, 0:768]
y = data['Average_Marks']

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Model and threshold
model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
threshold = 50

# Train and predict
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Regression metrics
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, y_pred)

# Convert predictions to binary classes based on threshold
y_pred_classes = (y_pred >= threshold).astype(int)
y_test_classes = (y_test >= threshold).astype(int)

# Classification metrics
accuracy = accuracy_score(y_test_classes, y_pred_classes)
precision = precision_score(y_test_classes, y_pred_classes, zero_division=0)
recall = recall_score(y_test_classes, y_pred_classes, zero_division=0)
f1 = f1_score(y_test_classes, y_pred_classes, zero_division=0)
conf_matrix = confusion_matrix(y_test_classes, y_pred_classes)

# Display regression metrics
print("Polynomial Regression (2nd Degree) Performance:")
print(f"MSE: {mse}, RMSE: {rmse}, RÂ²: {r2}")
print(f"Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1 Score: {f1}")

# Plot confusion matrix with a colorful palette
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="YlGnBu", cbar=True, annot_kws={"size": 14},
            xticklabels=["Predicted Negative", "Predicted Positive"], 
            yticklabels=["Actual Negative", "Actual Positive"],
            linewidths=1, linecolor='black', square=True)

plt.xlabel("Predicted Class")
plt.ylabel("Actual Class")
plt.title("Confusion Matrix with Colors")
plt.show()

# Plot classification metrics
classification_metrics = {
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall,
    "F1 Score": f1                                             
}

# Convert metrics dictionary to DataFrame for plotting
metrics_df = pd.DataFrame(list(classification_metrics.items()), columns=["Metric", "Score"])

# Plot the classification metrics
plt.figure(figsize=(8, 5))
sns.barplot(data=metrics_df, x="Metric", y="Score", palette="coolwarm")
plt.title("Classification Metrics for Polynomial Regression (2nd Degree)")
plt.ylabel("Score")
plt.ylim(0, 1)  # Assuming scores are within 0-1
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
